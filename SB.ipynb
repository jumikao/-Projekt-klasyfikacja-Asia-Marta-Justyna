# **WYKRES** **PUNKTOWY**

```python
sns.lmplot(x="Color intensity", y="Magnesium", data=wine)       #inna kolejność-najpierw definiujemy x,y, a potem do jakich danych się odnosimy data=wine
#lmplot- linear moodule - linia regresji (trendu)-generuje się automatycznie
```

```python
sns.lmplot(x="Alcohol", y="Magnesium", data=wine, fit_reg=False, hue="Class")  
#fit_reg=False usuwa linię regresji 
#hue="Class"- definiuje legende oraz zróżnicowane kolorowanie po kolumnie "class"
```

```python
sns.scatterplot(x="Color intensity", y="Magnesium", data=wine, hue= "Class")      #wykres punktowy
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x="Alcohol", y="Magnesium", data=wine, hue="Class")

#podpis na osi x,y
plt.xlabel("Zawartość alkoholu")            #różnica w pakiecie seaborn w stosunku do matplotlib!!
plt.ylabel("Zawartość magnezu")

#zmiana wartości na osi x,y
plt.xlim(0,15)
plt.ylim(0,170)

#tytuł
plt.title("Zawartość alkoholu i magnezu w różnych kategoriach wina")

#zmiana lokalizacji legendy
plt.legend(loc="upper left")


```

```python
sns.lmplot(x="Alcohol", y="Flavanoids", data=wine, fit_reg=False, hue="Class") 

```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x="Alcohol", y="Flavanoids", data=wine, hue="Class")

#podpis na osi x,y
plt.xlabel("Zawartość alkoholu")            #różnica w pakiecie seaborn w stosunku do matplotlib!!
plt.ylabel("Zawartość flawonoidów")

# #zmiana wartości na osi x,y
# plt.xlim(0,15)
# plt.ylim(0,170)

#tytuł
plt.title("Zawartość alkoholu i flawonoidów w różnych kategoriach wina")

#zmiana lokalizacji legendy
plt.legend(loc="upper left")
```

```python
plt.figure(figsize=(8,6))
sns.violinplot(x="Class",
               y="Alcohol",
               data=wine,
               inner=None)      #usuwamy czarną linię ze środka

sns.swarmplot(x="Class",
              y="Alcohol",
              data=wine,
              color="k",
              alpha=0.7)

plt.title("Zawartość alkoholu w zależności od wina")
```

```python
plt.figure(figsize=(8,6))
sns.swarmplot(x="Flavanoids", y="Color intensity", data=wine)
plt.title("Zależność między intensywnością koloru wina, a zawartością flawonoidów w winie")
```

```python
sns.violinplot(x="Alcohol", y="Color intensity", data=wine)
plt.title("Zależność między intensywnością koloru wina, a zawartością alkoholu w winie")
```

```python
melted_df=pd.melt(wine.loc[:,['Alcohol', 'Ash', 'Flavanoids', 'Color intensity', 'Class']],
                  id_vars='Class',        #kolumna class zostaje
                  var_name='Stat')        #pozostałe wartości stają się kategoriami w kolumnie "stat"
```

```python
melted_df.head()  
```

```python
plt.figure(figsize=(10,6))

sns.swarmplot(x='Stat', y='value', data=melted_df, hue='Class')

plt.legend(loc=2, bbox_to_anchor=(1,1))         #bbox to anchor- legenda poza obszarem wykres

plt.tight_layout()          #jak zapisuje się obrazek funkcja pilnuje, aby wszystkie wartości znalazły się na wykresie (obrazku)
plt.savefig('obrazek.png')
```
